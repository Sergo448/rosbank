```python
import pandas as pd
import numpy as np


# path to csv's files
path_to_history_prices = '..//csv_s//history_prices.csv'
path_to_investment_currencyhistory = '..//csv_s//investment_currencyhistory.csv'
path_to_securities = '..//csv_s//securities.csv'
path_to_сlients = '..//csv_s//history_prices.csv'

# read files
history_prices = pd.read_csv(path_to_history_prices)
investment_currencyhistory = pd.read_csv(path_to_investment_currencyhistory) 
securities = pd.read_csv(path_to_securities) 
сlients = pd.read_csv(path_to_сlients)

# drop column
history_prices = history_prices.drop(columns=['Unnamed: 0'])
investment_currencyhistory = investment_currencyhistory.drop(columns=['Unnamed: 0'])
securities = securities.drop(columns=['Unnamed: 0'])
securities['client_date'] = pd.to_datetime(securities['client_date'], format='%Y-%m-%d')

сlients = сlients.drop(columns=['Unnamed: 0'])

def informer(dataframe):
    
    print(f'INFORMATION FOR DATAFRAME:')
    print(dataframe.info())
    print('---------------')

    # Проверка на отсутствие каких-либо данных
    print('pd.ISNULL().SUM()')
    print(dataframe.isnull().sum())
    print('---------------')

    # Посчитаем количество уникальных гномов
    print(f'Количество уникальных значений для каждой колонки:\n'
          f'{dataframe.nunique()}')
    print('---------------')
    
# HISTORY_PRICES
print('HISTORY_PRICES')
print(informer(dataframe=history_prices))

# INVESTMENT_CURRENCYHISTORY
print('INVESTMENT_CURRENCYHOSTORY')
print(informer(dataframe=investment_currencyhistory))

# SECURITIES
print('SECURITIES')
print(informer(dataframe=securities))

# CLIENTS
print('CLIENTS')
print(informer(dataframe=сlients))
```

```python
currency_history_for_mean = investment_currencyhistory.copy()

currency_history_for_mean = currency_history_for_mean.drop(columns=['currency_id'])
currency_history_for_mean = currency_history_for_mean.sort_values(by=['currency', 'quote_date'])
currency_history_for_mean['quote_date'] = pd.to_datetime(currency_history_for_mean['quote_date'])
currency_history_for_mean['quote_date'] = currency_history_for_mean['quote_date'].dt.strftime('%Y-%m-%d')
currency_history_for_mean = currency_history_for_mean.reset_index()
currency_history_for_mean = currency_history_for_mean.rename(columns={"quote_date": "DATE"}, errors="raise")

currency_history_for_mean
```

```python
securities = securities.sort_values(by=['client_isin_paper', 'client_date'])

# Для начала посчитаем среднюю цену акции в ее валюте
# Для этого возьмем цену обшую в день на момент закрытия и количество той или иной акции
# Затем поделим и найдем среднюю цену

for_mean_table = pd.DataFrame()
for_mean_table['ISIN'] = securities['client_isin_paper']
for_mean_table['DATE'] = securities['client_date'].dt.strftime('%Y-%m-%d')
for_mean_table['AMOUNT'] = securities['client_n_papers']
for_mean_table['CURRENCY'] = securities['client_currency']
for_mean_table['COST'] = securities['client_closing_price']
for_mean_table = for_mean_table.reset_index()
for_mean_table['MEAN_DAILY'] = (for_mean_table ['COST'] / for_mean_table ['AMOUNT']).round(2)

for_mean_table
```

```python
# Далее нам надо найти средневзвешенное среднее
# Для этого у нас имеется датафрейм с данными по ценам торгования валют 
# Нам надо также перевевсти все в какую-то одну валюту, пусть опять это будет доллар
# Нам надо найти как торгуется та или иная валюта в этот/ближайший день к закрытию рынка
# Перевести наше полученное уже среднее в эту валюту
# И найти новое среднее по получившимся данным


# Способ наверняка. Объединение таблиц по датам.

res_t = for_mean_table.merge(currency_history_for_mean, how='left', on='DATE')
res_t = res_t.dropna()
res_t = res_t.drop(columns=['index_x', 'index_y'])
res_t = res_t.sort_values(by=['CURRENCY', 'currency']).reset_index(inplace=False)
res_t = res_t.drop(columns=['index'])
```

```python
res_t 
```

```python
cols_for_drop = ['DATE', 'AMOUNT', 'COST', 'reverse_quote', 'cross_course']
res_t_RUB = pd.DataFrame()
res_t_AUD = pd.DataFrame()
res_t_EUR = pd.DataFrame()
res_t_USD = pd.DataFrame()

"""
res_t_RUB = res_t.copy()
res_t_AUD = res_t.copy()
res_t_EUR = res_t.copy()
res_t_USD = res_t.copy()
"""

"""
res_t_RUB = res_t_RUB.drop(columns=cols_for_drop)
res_t_AUD = res_t_AUD.drop(columns=cols_for_drop)
res_t_EUR = res_t_EUR.drop(columns=cols_for_drop)
res_t_USD = res_t_USD.drop(columns=cols_for_drop)
"""

res_t_RUB['in_USD'] = res_t[(res_t['currency'] == 'RUB/USD')]['MEAN_DAILY'] / res_t['direct_quote']
res_t_AUD['in_USD'] = res_t[(res_t['currency'] == 'AUD/USD')]['MEAN_DAILY'] / res_t['direct_quote']
res_t_EUR['in_USD'] = res_t[(res_t['currency'] == 'EUR/USD')]['MEAN_DAILY'] / res_t['direct_quote']
res_t_USD['in_USD'] = res_t[(res_t['currency'] == 'USD/USD')]['MEAN_DAILY'] / 1

r_t = pd.DataFrame()

r_t['MEAN_DAILY_IN_USD'] = pd.concat([res_t_RUB['in_USD'], res_t_AUD['in_USD'],
                                        res_t_EUR['in_USD'], res_t_USD['in_USD']],
                                        ignore_index=True)
r_t['ISIN'] = res_t['ISIN']

del(res_t_RUB, res_t_AUD, res_t_EUR, res_t_USD)


r_t = r_t.dropna().reset_index(inplace=False).drop(columns=['index']).sort_values(by=['ISIN'])

res_list = r_t.values.tolist()

del(res_t)
```

```python
means = []
buf = []
for i in range(len(res_list)):

    try: 
        if res_list[i][1] == res_list[i+1][1]:
            buf.append(res_list[i][0])
        else:
            buf.append(res_list[i][0])
            mean = np.mean(buf)
            means.append(mean)
            buf = []
    except IndexError:
        buf.append(res_list[i][0])
        mean = np.mean(buf)
        means.append(mean)
        buf = []  
```

```python
ISIN_LIST = r_t['ISIN'].unique()

del(r_t)

result_for_mean_with_weights = pd.DataFrame(data = {
    'ISIN' : ISIN_LIST,
    'MEAN' : means
})
```

## Таким образом мы получили среднее значение для каждой бумаги

В ходе решения была найденна средняя цена для каждой бумаги в день закрытия рынка в валюте в которой она была продана/куплена. Затем было выполнено сопоставление по датам. Затем был выбран подходящий курс для перевода всех средних в одну валюту (были выбраны доллары, при жедании может быть любая другая валюта). Затем после переведения всего в одну валюту было вычисленно финальное сревдневзвешенное среднее. Несмотря на небольшьшие принебрежения, способ был выдержан. При необходимости можно масштабировать, усложнить и автоматизировать код.
